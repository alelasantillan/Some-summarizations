```python
#Now split text in an array with less than 2500 tokens each:
n = 2500
n = 1500

paragraphs = iter(text.split(".\n"))
lines, current = [], next(paragraphs)
for paragraph in paragraphs:
    paragraph = paragraph.rstrip('\n')
    if len(current) + 2 + len(paragraph) > n:
        #current = current.rstrip('\n')
        lines.append(current)
        current = paragraph + '.'
    else:
        current += " " + paragraph + '.'
lines.append(current)
```

```python
for j in range(len(lines)):
  print(j,len(lines[j]),lines[j],"\n")
```

```python
#finally we summarize:
for j in range(len(lines)):
  res = summarizer(lines[j], min_length=50, max_length=80)
  print(j,len(res[0]['summary_text']),res[0]['summary_text'])
```
